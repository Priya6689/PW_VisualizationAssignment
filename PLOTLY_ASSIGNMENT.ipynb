```python
#1. Using the given dataset, to generate a 3D scatter plot to visualize the distribution of data points in a three- dimensional space.
#np.random.seed (30)
#data =
#'X': np.random.uniform(-10, 10, 300), Y': np.random.uniform(-10, 10, 300),
#' 'Z': np.random.uniform(-10, 10, 300)
#} df = pd.DataFrame(data)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
# Seed for reproducibility
np.random.seed(30)

# Create the dataset
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}
df = pd.DataFrame(data)

# Create a 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the data points
ax.scatter(df['X'], df['Y'], df['Z'], c='blue', marker='o', alpha=0.6)

# Set labels and title
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
ax.set_title('3D Scatter Plot of Data Points')

# Show the plot
plt.show()
```

```python
#2. Using the Student Grades, create a violin plot to display the distribution of scores across different grade categories.
#np.random.seed (15)
#data = {
#'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
#'Score': np.random.randint(50, 100, 200)
#}
#df = pd.DataFrame (data)
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
# Set seed for reproducibility
np.random.seed(15)

# Generate the data
data = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create a violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='Grade', y='Score', data=df, inner='quartile', palette='muted')

# Adding labels and title
plt.xlabel('Grade')
plt.ylabel('Score')
plt.title('Distribution of Scores Across Different Grade Categories')

# Show the plot
plt.grid(True)
plt.show()
```

```python
#3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.
#np.random.seed (20)
#data = {
#'Month': np.random.choice (['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
#'Day': np.random.choice(range (1, 31), 100),
#'Sales': np.random.randint(1000, 5000, 100)
#}
#df = pd.DataFrame (data)
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(20)

# Generate the data
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create a pivot table for the heatmap
pivot_table = df.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='mean')

# Create a heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(pivot_table, annot=True, fmt=".1f", cmap='YlGnBu', cbar_kws={'label': 'Average Sales'})

# Adding labels and title
plt.xlabel('Month')
plt.ylabel('Day of the Month')
plt.title('Heatmap of Sales Variation Across Different Months and Days')

# Show the plot
plt.show()
```

```python
#4. Using the given x and y data, generate a 3D surface plot to visualize the function z = sin(x²+y²
#x = np.linspace (-5, 5, 100)
#y = np.linspace (-5, 5, 100)
#x, y = np.meshgrid(x, y)
#z = np.sin(np.sqrt(x**2 + y**2))
#data = {
#'X': x.flatten(),
#'Y': y.flatten(),
#'Z': z.flatten()
#}
#df = pd.DataFrame (data)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate the data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create a DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create a 3D surface plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plotting the surface
ax.plot_surface(df['X'].values.reshape(100, 100), df['Y'].values.reshape(100, 100),
                 df['Z'].values.reshape(100, 100), cmap='viridis')

# Adding labels and title
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis (sin(sqrt(x² + y²)))')
ax.set_title('3D Surface Plot of z = sin(sqrt(x² + y²))')

# Show the plot
plt.show()
```

```python
#5. Using the given dataset, create a bubble chart to represent each country's population (y-axis), GDP (x- axis), and bubble size proportional to the population.
#np.random.seed(25)
#data = {
#'Country': ['USA', 'Canada', 'UK',
#'Germany', 'France'],
#'Population':
#np.random.randint(100, 1000, 5),
#'GDP': np.random.randint(500, 2000,
#5)
#}
#df pd.DataFrame (data)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(25)

# Generate the dataset
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create a bubble chart
plt.figure(figsize=(10, 6))
plt.scatter(df['GDP'], df['Population'], s=df['Population']*2, alpha=0.5, c='blue', edgecolors='w', linewidth=2)

# Adding labels and title
for i in range(len(df)):
    plt.annotate(df['Country'][i], (df['GDP'][i], df['Population'][i]), fontsize=10, ha='center', va='center')

plt.xlabel('GDP (in billions)')
plt.ylabel('Population (in millions)')
plt.title('Bubble Chart of Population vs. GDP by Country')

# Show the plot
plt.grid(True)
plt.xlim(400, 2100)  # Adjust x-axis limits
plt.ylim(0, 1000)    # Adjust y-axis limits
plt.show()
```
