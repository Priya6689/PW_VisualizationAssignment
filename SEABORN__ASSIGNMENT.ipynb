```python
#1. Create a scatter plot to visualize the relationship between two variables, by generating a synthetic dataset.
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Generate synthetic dataset
np.random.seed(42)
data = pd.DataFrame({
    'Variable X': np.random.randn(100),
    'Variable Y': np.random.randn(100) + np.random.randn(100) * 0.5
})

# Create a scatter plot
sns.scatterplot(x='Variable X', y='Variable Y', data=data)

# Add labels and title
plt.title('Scatter Plot of Variable X vs Variable Y')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')

# Display the plot
plt.show()
```

```python
#2. Generate a dataset of random numbers. Visualize the distribution of a numerical variable.
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Generate a random dataset
data = np.random.randn(1000)

# Create a distribution plot
sns.histplot(data, kde=True, color='skyblue', bins=30)

# Add labels and title
plt.title('Distribution of Random Numbers')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

```python
#3. Create a dataset representing categories and their corresponding values. Compare different categories based on numerical values.
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)

# Create a dataset with categories and values
data = pd.DataFrame({
    'Category': ['Category A', 'Category B', 'Category C', 'Category D', 'Category E'],
    'Value': [25, 40, 30, 35, 20]
})

# Create a bar plot
sns.barplot(x='Category', y='Value', data=data, palette='pastel')

# Add labels and title
plt.title('Comparison of Categories Based on Values')
plt.xlabel('Category')
plt.ylabel('Value')

# Display the plot
plt.show()
```

```python
#4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories.
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
# Create a dataset with categories and numerical values
np.random.seed(42)  # For reproducibility
data = pd.DataFrame({
    'Category': np.random.choice(['Category A', 'Category B', 'Category C'], size=300),
    'Value': np.random.randn(300) * 10 + 50  # Random values centered around 50
})

# Create a box plot with hue
sns.boxplot(x='Category', y='Value', data=data, hue='Category', palette='pastel', legend=False)

# Add labels and title
plt.title('Distribution of Values Across Categories')
plt.xlabel('Category')
plt.ylabel('Value')

# Display the plot
plt.show()
```

```python
#5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate a synthetic dataset with correlated features
np.random.seed(42)  # For reproducibility
n_samples = 100

# Create correlated features
x = np.random.randn(n_samples)
y = 2 * x + np.random.randn(n_samples) * 0.1  # y is correlated with x
z = 0.5 * x + 0.5 * y + np.random.randn(n_samples) * 0.1  # z is correlated with both x and y

# Combine into a DataFrame
data = pd.DataFrame({'Feature X': x, 'Feature Y': y, 'Feature Z': z})

# Calculate the correlation matrix
correlation_matrix = data.corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True, cbar_kws={"shrink": .8})

# Add title
plt.title('Correlation Matrix Heatmap')

# Display the plot
plt.show()
```
